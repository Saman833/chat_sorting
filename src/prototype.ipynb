```python
def show_component(root,vertex_list:list):
    def filter_node(node):
        return node.replace("message_","")
    for vertex in vertex_list:
        print(filter_node(vertex))
    root=filter_node(root)
    import networkx as nx
    import matplotlib.pyplot as plt
    from collections import deque
    
    def get_edges()-> list:
        edges = []
        for vertex in vertex_list:
            edges.extend([(filter_node(str(vertex)), filter_node(str(v)),w) for v,w in graph.v_neighbors[vertex]])
        print(edges)
        return edges
    def bfs_layered_layout(G, root):
        """
        Return a dict of positions for a BFS-layered layout of a directed graph.
        Ignores any edge weights/labels (pure BFS layering).
        """
        from collections import deque
        
        queue = deque([(root, 0)])
        visited = {root}
        levels = {}
        
        while queue:
            node, level = queue.popleft()
            levels.setdefault(level, []).append(node)
            
            for successor in G.successors(node):
                if successor not in visited:
                    visited.add(successor)
                    queue.append((successor, level + 3))
        
        pos = {}
        for level_idx in sorted(levels.keys()):
            nodes_on_level = levels[level_idx]
            for i, node in enumerate(nodes_on_level):
                pos[node] = (i, -2*level_idx)
        
        return pos
    # Example usage

    # 1) Create a directed tree with edges
    root=filter_node(root)
    G = nx.DiGraph()
    # for node in vertex_list:
    #     G.add_node(filter_node(node))
    edges =get_edges()
    for edge in edges:
        G.add_edge(edge[0],edge[1],label=edge[2].replace("user","")[:3])

    pos = bfs_layered_layout(G, root)

    # 3) Plot the graph
    plt.figure()
    nx.draw_networkx(
        G, 
        pos, 
        arrows=True,  
        with_labels=True,      # Show arrows for a DiGraph
          # Display node labels
        node_size=500,
        arrowstyle='-|>',   # A nicer arrow style
        arrowsize=15
    )
    edge_labels = nx.get_edge_attributes(G, 'label')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    plt.axis('off')
    plt.show()

```
